```python
import folium
import panel as pn

pn.extension(sizing_mode="stretch_width")
```

```python
m = folium.Map(location=[48.86598, 2.275725], zoom_start=12)

folium_pane = pn.pane.plot.Folium(m, height=600)

folium_pane
```

```python
# Add a marker to the map
folium.Circle(
    [48.86598, 2.275725], popup="<i>Brandenburg Gate</i>", tooltip="Click me!", fill=0.9
).add_to(m)

folium_pane.object = m
```
